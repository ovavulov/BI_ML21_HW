# SVM и его ядра
__Суммарное количество баллов: 10__

```python
import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, average_precision_score, roc_curve, precision_recall_curve, classification_report
from mlxtend.plotting import plot_decision_regions
from sklearn.ensemble import RandomForestClassifier

SEED = 19
```

```python
def make_classification_report(y_true, y_pred, y_proba):
    print(classification_report(y_true, y_pred))

    f, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
    fpr, tpr, thresholds = roc_curve(y_true, y_proba)
    roc_score = roc_auc_score(y_true, y_proba)
    ax[0].set_title(f"ROC curve (AUC: {roc_score:.4f})", fontsize=15)
    ax[0].grid()
    ax[0].plot(fpr, tpr)

    fpr, tpr, thresholds = precision_recall_curve(y_true, y_proba)
    ap_score = average_precision_score(y_true, y_proba)
    ax[1].set_title(f"PR curve (AUC: {ap_score:.4f})", fontsize=15)
    ax[1].grid()
    ax[1].plot(fpr, tpr);
```

### Задание 1 (7 баллов)

Примените метод SVM (например, из библиотеки sklearn) для датасета blobs2.

Визуализируйте результат (разбиение плоскости и опорные вектора) при разных вариантах ядер (линейное; полиномиальное степеней 2,3,5; RBF).

```python
data = pd.read_csv("./hw7_data/blobs2.csv")
X, y = data.drop("label", axis=1), data.label
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)
```

```python
model = SVC(kernel="linear").fit(X_train, y_train)
plt.figure(figsize=(10, 8))
plt.title("SVM result for linear kernel", fontsize=20)
sv = model.support_vectors_
plot_decision_regions(X.values, y.values, clf=model, legend=2, X_highlight=sv);
```

```python
model = SVC(kernel="poly", degree=2).fit(X_train, y_train)
plt.figure(figsize=(10, 8))
plt.title("SVM result for polynomial kernel (degree = 2)", fontsize=20)
sv = model.support_vectors_
plot_decision_regions(X.values, y.values, clf=model, legend=2, X_highlight=sv);
```

```python
model = SVC(kernel="poly", degree=3).fit(X_train, y_train)
plt.figure(figsize=(10, 8))
plt.title("SVM result for polynomial kernel (degree = 3)", fontsize=20)
sv = model.support_vectors_
plot_decision_regions(X.values, y.values, clf=model, legend=2, X_highlight=sv);
```

```python
model = SVC(kernel="poly", degree=5).fit(X_train, y_train)
plt.figure(figsize=(10, 8))
plt.title("SVM result for polynomial kernel (degree = 5)", fontsize=20)
sv = model.support_vectors_
plot_decision_regions(X.values, y.values, clf=model, legend=2, X_highlight=sv);
```

```python
model = SVC(kernel="rbf").fit(X_train, y_train)
plt.figure(figsize=(10, 8))
plt.title("SVM result for RBF kernel", fontsize=20)
sv = model.support_vectors_
plot_decision_regions(X.values, y.values, clf=model, legend=2, X_highlight=sv);
```

### Задание 2 (3 балла)

Примените метод SVM на датасетах cancer и spam, разбитый предварительно на train (80%) и validate (20%). 

Сравните время работы SVM и Random Forest при близкой точности.

```python
home_path = os.getcwd()
CANCER_DATA_PATH = os.path.join(home_path, "hw2_data", "cancer.csv")
SPAM_DATA_PATH = os.path.join(home_path, "hw2_data", "spam.csv")
```

```python
def read_cancer_dataset(path_to_csv):
    df = pd.read_csv(path_to_csv)
    X, y = df.drop("label", axis=1), df.label
    return X, y

def read_spam_dataset(path_to_csv):
    return read_cancer_dataset(path_to_csv)
```

```python
X_cancer, y_cancer = read_cancer_dataset(CANCER_DATA_PATH)
y_cancer = y_cancer.map({"M": 0, "B": 1})
X_spam, y_spam = read_spam_dataset(SPAM_DATA_PATH)
```

# Cancer

```python
np.random.seed(SEED)
X_train, X_test, y_train, y_test = train_test_split(X_cancer, y_cancer, test_size=0.2)
```

SVM

```python
%%timeit
svm = SVC(probability=True).fit(X_train, y_train)
y_pred = svm.predict(X_test)
y_proba = svm.predict_proba(X_test)[:, 1]
```

```python
svm = SVC(probability=True).fit(X_train, y_train)
y_pred = svm.predict(X_test)
y_proba = svm.predict_proba(X_test)[:, 1]
```

```python
make_classification_report(y_test, y_pred, y_proba)
```

Random Forest

```python
%%timeit
rf = RandomForestClassifier().fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]
```

```python
rf = RandomForestClassifier().fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]
```

```python
make_classification_report(y_test, y_pred, y_proba)
```
